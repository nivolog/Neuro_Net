## 1) Анализ данных
Так как показатели мгновенных ускорений гироскопа и акселлерометра даны в мировой системе координат, т.е. в системе, где мы можем предполагать, что ускорения направлены по осям движения относительно человека. То было принято решение не производить манипуляций со значениями столбцов, а просто "склеить" показатели приборов в одну таблицу с временнымм метками.
## 2) Модель
Изучив приведённую статью, было принято решение использовать LSTM нейросеть с параметрами…., ТК это рекуррентная нейросеть хорошо подходит для задачи регрессии с временными рядами
## 3) Результаты
Решение состоит из двух файлов: LSTM_Net и Prediction, первый файл создает модель и обучает ее, второй - использует для предсказания траекторий и создает файлы траекторий

Полученные результаты указывают на переобученность модели (среднее квадратичное отклонение равно полоовине максимума амплитуды координат) и неоптимальные параметры.
